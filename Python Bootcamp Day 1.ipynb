# Variables in Python

In python, a variable is seen as tag that is mapped to some value. 

```python
num=14
```

Python considers the value i.e., 14 in the above case as objects. If the value of variable num is changed to new value then the tag simply changed to the new value as shown below. The value 14 will be unreferenced, it will be removed by garbage collector.

```python
num=20
new_num=num
```

Assigning one variable to another variable makes a new tag bound to the same value. In this case there is only memory is allocated for two variables with two same values. Byt his way Python is using <b>memory efficiently</b>.

# id() function

id() is an inbuilt function in Python.
id function accepts a single parameter and is used to return the identity of an object.
This identity has to be unique and constant for this object during the lifetime.If we relate this to C, then they are actually the memory address, here in Python it is the unique id.

```python
#let's check the id of num and new_num variables
num_id=id(num)
new_num_id=id(new_num)
print(num_id)
print(new_num_id)
```

## String - formatting methods

### .format method

The format() method formats the specified value(s) and insert them inside the string's placeholder.

The placeholder is defined using curly brackets: {}.

```python
language=input("Enter any programming language:")
mystr="Welcome to {} Bootcamp".format(language)
print(mystr)
```

```python
# if there are n variables to be inserted in the string then the varibles are given as parameters in a order to be inserted
language=input("Enter any programming language:")
date=input("Enter the date of Bootcamp:")
mystr="Welcome to {} Bootcamp on {}".format(language,date)
print(mystr)
```

```python
# If we reverse the parameters of format() then the the variables also placed in wrong places
language=input("Enter any programming language:")
date=input("Enter the date of Bootcamp:")
mystr="Welcome to {} Bootcamp on {}".format(date,language)
print(mystr)
```

```python
# To remove confusion we will introduce variables inside the curly brackets, now order of parameters doesn't matter
language=input("Enter any programming language:")
date=input("Enter the date of Bootcamp:")
mystr="Welcome to {lang} Bootcamp on {when}".format(when=date,lang=language)
print(mystr)
```
